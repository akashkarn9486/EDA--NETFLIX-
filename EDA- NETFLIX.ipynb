<b>CODE BLOCK DEMONSTRATING BASIC INFO ABOUT THE DATASET .</b>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
file_path = "Netflix Engagement Dataset.xlsx"
df = pd.read_excel(file_path, sheet_name="Netflix Engagement Dataset")

# Set aesthetic style
sns.set(style="whitegrid")

# Show basic info
print("Basic Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe(include="all"))
```

<b>PLOT 1:-</b><br> DEMONSTRATING THE DISTRIBUTION OF CUSTOMER SATISFACTION SCORE 
   THIS PLOT PARTICULARLY SHOWCASES THE GRAPH IN THE FORM OF A HISTOGRAM . THIS PLOT IS TITLED AS DISTRIBUTION OF CUSTOMER SATISFACTION SCORE . WITH THE X AXIS LABELLED AS SCORE(1-10) . WITH THE Y AXIS LABELLED AS FREQUENCY . THIS PLOT CAN MAKE THE USER UNDERSTAND THE CUSTOMER SATISFACTION SCORE BASED ON THE DATASET.

```python
plt.figure(figsize=(8, 4))
sns.histplot(df["Customer Satisfaction Score (1-10)"], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Customer Satisfaction Score")
plt.xlabel("Score (1-10)")
plt.ylabel("Frequency")
plt.show()
```

<b>PLOT2:-</b><br> THIS PLOT SHOWCASES THE BOXPLOT OF DAILY WATCH TIME AGAINST THE SUBSCRIPTION PLAN OPTED BY THE USER. THIS PLOT PARTICULARLY SIGNIFIES THE VARIATION OF WATCH TIME BASED ON THE PLAN OPTED .

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Subscription Plan", y="Daily Watch Time (Hours)", palette="pastel")
plt.title("Daily Watch Time by Subscription Plan")
plt.show()
```

<b>PLOT3:-</b><br> THIS SCATTER PLOT PARTICULARLY PLOTS THE CUSTOMER SATISFACTION SCORE AND THE ENGAGEMENT RATE BASED ON THE DATASET PROVIDED .

```python
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df, x="Customer Satisfaction Score (1-10)", y="Engagement Rate (1-10)", hue="Churn Status (Yes/No)", palette="Set1")
plt.title("Engagement vs Satisfaction by Churn Status")
plt.show()
```

<b>PLOT 4:-</b><br> THIS PLOT PARTICULARLY SIGNIFIES THE GENRE PREFERENCE OF THE INDIVIDUALS USING NETFLIX AS THEIR OTT SITE.

```python
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x="Genre Preference", order=df["Genre Preference"].value_counts().index, palette="muted")
plt.title("Genre Preferences")
plt.xticks(rotation=45)
plt.show()
```

<b>PLOT 5:-</b><br>THIS PLOT DISPLAYS THE HEATMAP OF THE CORRELATION MATRIX FORMED BY THE METRICES ('CUSTOMER ID','SUBSCRIPTION LENGTH','CUSTOMER SATISFACTION SCORE','DAILY WATCH TIME','ENGAGEMENT RATE','SUPPORT QUERIES LOGGED','AGE','MONTHLY INCOME','NO OF PROFILES CREATED'.

```python
# Plot: Heatmap of Correlation Matrix
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

<b> PLOT 6:-</b><br> THIS PLOT SHOWCASES THE SCATTERPLOT BETWEEN MONTHLY INCOME AND DAILY WATCH TIME BASED ON THE DEVICES USED .

```python
# Plot: Monthly Income vs Daily Watch Time with Device Used
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Monthly Income ($)", y="Daily Watch Time (Hours)", hue="Device Used Most Often", palette="Set2")
plt.title("Income vs Watch Time by Device")
plt.show()
```

<b>PLOT 7:-</b><br> THIS PLOT SHOWS AGE DISTRIBUTION BY CHURN STATUS 

```python
# Plot: Age distribution segmented by Churn Status
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="Age", hue="Churn Status (Yes/No)", bins=15, kde=True, palette="Set1")
plt.title("Age Distribution by Churn Status")
plt.show()
```
